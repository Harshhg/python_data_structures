```python
def findShortest(graph):
    dist =  [0]*N
    dist[N-1]=0
    i=6
    while(i>=0):
        dist[i]=INF
        
        for j in range(i+1,N):
            if graph[i][j] != INF :
                minvalue = min(dist[i], graph[i][j]+dist[j])
                dist[i]=minvalue
        i-=1
        
    # Returning the first value, since it contains the minimum calculated cost.
    return dist[0]
```

```python
N=8
INF=999999
graph = [[INF, 1, 2, 5, INF, INF, INF, INF],  
         [INF, INF, INF, INF, 4, 11, INF, INF],  
         [INF, INF, INF, INF, 9, 5, 16, INF],  
         [INF, INF, INF, INF, INF, INF, 2, INF],  
         [INF, INF, INF, INF, INF, INF, INF, 18], 
         [INF, INF, INF, INF, INF, INF, INF, 13],  
         [INF, INF, INF, INF, INF, INF, INF, 2]]  
  
```

```python
print(findShortest(graph))
```
